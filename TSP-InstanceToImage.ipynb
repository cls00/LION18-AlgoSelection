```python
## Generate All Images
directory = "/content/GREEDY"
imgdir = "/content/GRDimages"
for filename in os.listdir(directory):
    instance = os.path.join(directory, filename)
    df_tsp_instance = pd.read_csv(instance,names=["x","y"])
    fig = figure(figsize=(4, 4), dpi=8)
    plt.style.use('grayscale')
    plt.scatter(df_tsp_instance['x'],df_tsp_instance['y'],s=30)
    plt.axis('off')
    image_file = os.path.join(imgdir, filename[:-4])
    plt.savefig(image_file+".png", dpi=8 )
    #imgplot = plt.imshow(img)
    #plt.imsave(image_file+".png", dpi=8 )
    plt.close()

```
